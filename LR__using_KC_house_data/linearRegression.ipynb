```python
#import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Once these libraries have been imported our next step will be fetching the dataset and loading our data. While loading the data, it is important to note that the file name shall have its format (.csv/.xls) specified at the end.

```python
#Read the Dataset
df = pd.read_csv('kc_house_data.csv')
```

STEP 2: VISUALISING THE DATA 
You need to visualize the data, to know which parameters you will take to build the model.
Matplotlib and Seashore are excellent libraries that can be used to visualize our data on various different plots

```python
#visualizing data using heatmap
plt.figure()
sns.heatmap(df.corr(),cmap='coolwarm')
plt.show()
```

STEP 3: FEATURE ENGINEERING
Only by visualzing the data we can find if there exist any parameters that are strongly related.

```python
#Selecting the required parameters
area = df['sqft_living']
price = df['price']

x = np.array(area).reshape(-1,1)
y = np.array(price)
```

4. Fitting Linear Regression Model
Next lets import train_test_split method and linear Regression Model from sci-kit learn library and train the model with our data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state = 0)

y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)

#Fit the model over the training dataset
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#calculate intercept and cofficient
print(model.intercept_)
print(model.coef_)

pred = model.predict(x_test)
predictions = pred.reshape(-1,1)

#calculate root mean squared error to evaluate model performance
from sklearn.metrics import mean_squared_error
print('MSE : ',mean_squared_error(y_test,predictions))
print('RMSE : ',np.sqrt(mean_squared_error(y_test,predictions)))
```
